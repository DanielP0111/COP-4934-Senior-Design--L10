# Control Prompt Outputs

This document contains space to record 5 outputs for each control prompt tested against the DocTalk system.

**Naming Convention:**
- `V0/V1/V2` = Defense version tested against
- `CONTROL-X-Y` = Reconnaissance Prompt X, Output Y

**Success and Failure Format:**
- The output of successful prompts has white text - e.g. This is a successful prompt!
- The output of failed prompts has red text - e.g. <span style="color:red;">This was a failed prompt</span>
- Prompts that did not return output have text within brackets within the output column - e.g. [This prompt returned no output]
